# Múltiples menus con volver y salir

Vamos a modelar una situación donde tenemos 4 menús distintos que pueden llamar distintas acciones:
- **Menú 1**: menú 2 o menú 3.
- **Menú 2**: imprimir un mensaje o menú 4.
- **Menú 3**: imprimir un mensaje.
- **Menú 4**: imprimir un mensaje.

Todos los menús tienen la opción de **salir** y los menús 2, 3 y 4 pueden **volver atrás**. 

## Método 1 - Recursión **controlada**

Haremos que cada menú pueda llamar al siguiente menú, pero cada vez que haga salir o volver atrás. Haremos `returns`.

```python
VOLVER = "volver"
SALIR = "salir"

def menu_1():
    retorno_menu = None
    while retorno_menu != SALIR:
        print("""
        ### Soy menú 1 ###
        [a] Ir al menú 2
        [b] Ir al menú 3
        [c] Salir
        """)
        resp = input("¿Qué quieres hacer? (a, b o c): ")
        retorno_menu = 0
        if resp == "a":
            retorno_menu = menu_2()

        elif resp == "b":
            retorno_menu = menu_3()

        elif resp == "c":
            retorno_menu = SALIR

def menu_2():
    while True:
        print("""
        ### Soy menú 2 ###
        [a] Imprimir un mensaje
        [b] Ir al menú 4
        [c] Volver
        [d] Salir
        """)
        resp = input("¿Qué quieres hacer? (a, b, c o d): ")
        if resp == "a":
            print("\n\tola k ace")

        if resp == "b":
            retorno_menu = menu_4()
            if retorno_menu == SALIR:
                return SALIR

        elif resp == "c":
            return VOLVER

        elif resp == "d":
            return SALIR


def menu_3():
    while True:
        print("""
        ### Soy menú 3 ###
        [a] Imprimir otro mensaje
        [b] Volver
        [c] Salir
        """)
        resp = input("¿Qué quieres hacer? (a, b o c): ")
        if resp == "a":
            print("\n\tno se me ocurre un mensaje")

        elif resp == "b":
            return VOLVER

        elif resp == "c":
            return SALIR


def menu_4():
    while True:
        print("""
        ### Soy menú 4 ###
        [a] Imprimir otro mensaje
        [b] Volver
        [c] Salir
        """)
        resp = input("¿Qué quieres hacer? (a, b o c): ")
        if resp == "a":
            print("\n\tola k ace desde el menú 4")

        elif resp == "b":
            return VOLVER

        elif resp == "c":
            return SALIR
```

```python
menu_1()
```

##### ¿Notaste que cuando elegimos la opción "volver" o "salir" hacemos `return`?

Cuando elegimos la opción de "volver", por ejemplo en el menú 4, no llamamos nuevamente a `menu_2`, sino que hacemos `return` y luego, dentro de menú 2, se verifica el valor retornado por `menu_4` y toma una decisión al respecto. 

Si menú 4 dijo `SALIR` entonces el menú 2 también hace `return SALIR` para indicarle al `menu_1` que es hora de salir. Si menú 4 dijo otra cosa (`VOLVER` en este caso), menú 2 mantiene su _loop_ infinito hasta que digamos volver o salir.

## Método 2 - Menú manager

En esta ocasión, haremos una función principal encargada de administrar todos los menús, y en función de la opción que a ellos les ingresan, esta "administrador" llamará a los demás menús. Así evitamos que un menú llame a otro.

```python
VOLVER = "volver"
SALIR = "salir"
            
def menu_1():
    while True:
        print("""
        ### Soy menú 1 ###
        [a] Ir al menú 2
        [b] Ir al menú 3
        [c] Salir
        """)
        resp = input("¿Qué quieres hacer? (a, b o c): ")
        retorno_menu = 0
        if resp == "a":
            return "ir_menu_2"

        elif resp == "b":
            return "ir_menu_3"

        elif resp == "c":
            return SALIR


def menu_2():
    while True:
        print("""
        ### Soy menú 2 ###
        [a] Imprimir un mensaje
        [b] Ir al menú 4
        [c] Volver
        [d] Salir
        """)
        resp = input("¿Qué quieres hacer? (a, b, c o d): ")
        if resp == "a":
            print("\n\tola k ace")

        if resp == "b":
            return "ir_menu_4"

        elif resp == "c":
            return VOLVER

        elif resp == "d":
            return SALIR


def menu_3():
    while True:
        print("""
        ### Soy menú 3 ###
        [a] Imprimir otro mensaje
        [b] Volver
        [c] Salir
        """)
        resp = input("¿Qué quieres hacer? (a, b o c): ")
        if resp == "a":
            print("\n\tno se me ocurre un mensaje")

        elif resp == "b":
            return VOLVER

        elif resp == "c":
            return SALIR


def menu_4():
    while True:
        print("""
        ### Soy menú 4 ###
        [a] Imprimir otro mensaje
        [b] Volver
        [c] Salir
        """)
        resp = input("¿Qué quieres hacer? (a, b o c): ")
        if resp == "a":
            print("\n\tola k ace desde el menú 4")

        elif resp == "b":
            return VOLVER

        elif resp == "c":
            return SALIR


        
```

```python
def menu_manager():
    retorno_menu = None
    historial = ["ir_menu_1"]
    respuesta = ""
    while respuesta != SALIR:
        estado_actual = historial[-1]

        if estado_actual == "ir_menu_1":
            respuesta = menu_1()
        
        elif estado_actual == "ir_menu_2":
            respuesta = menu_2()
            
        elif estado_actual == "ir_menu_3":
            respuesta = menu_3()
            
        if respuesta == VOLVER:
            historial.pop()
            
        elif respuesta.startswith("ir"):
            historial.append(respuesta)
        
```
